```python
import os

root = "Sepsis/training_setA"
path = os.path.join(root, "training")

filenames = []
for path, subdirs, files in os.walk(root):
    for name in files:
        filenames.append(str(os.path.join(path, name)))
        #print(os.path.join(path, name))
        
filenames
```

```python
# load pandas
import pandas as pd

# read file as pandas dataframe
file = "Sepsis/training_setA/training/p000001.psv"
dataframe = pd.read_csv(file,sep='|')
```

```python
combined_csv = pd.concat( [ pd.read_csv(f, sep='|') for f in filenames ] )
```
