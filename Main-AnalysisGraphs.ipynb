```python
import altair as alt
Stat_Parity = alt.Chart(reweight_graph_df).mark_circle(size = circle_size).encode(
    alt.X("Accuracy_Tradeoff",
        scale=x_scale
    ),
    alt.Y('Stat_Parity_Diff',
        scale=y_scale
    ),
    color = "Dataset"
)

line = alt.Chart(pd.DataFrame({'Stat_Parity_Diff': [0]})).mark_rule().encode(y='Stat_Parity_Diff')
line2 = alt.Chart(pd.DataFrame({'Accuracy_Tradeoff': [0]})).mark_rule().encode(x='Accuracy_Tradeoff')

Stat_Parity = Stat_Parity + line + line2
Stat_Parity
```

```python
import altair as alt
Avg_Odd = alt.Chart(reweight_graph_df).mark_circle(size = circle_size).encode(
    alt.X("Accuracy_Tradeoff",
        scale=x_scale
    ),
    alt.Y('Avg_Odd_Diff',
        scale=y_scale
    ),
    color = "Dataset"
)

line = alt.Chart(pd.DataFrame({'Avg_Odd_Diff': [0]})).mark_rule().encode(y='Avg_Odd_Diff')
line2 = alt.Chart(pd.DataFrame({'Accuracy_Tradeoff': [0]})).mark_rule().encode(x='Accuracy_Tradeoff')
Avg_Odd = Avg_Odd + line + line2
Avg_Odd
```

```python
import altair as alt
Equal_Opp = alt.Chart(reweight_graph_df).mark_circle(size = circle_size).encode(
    alt.X("Accuracy_Tradeoff",
        scale=x_scale
    ),
    alt.Y('Equal_Opp_Diff',
        scale=y_scale
    ),
    color = "Dataset"
)

line = alt.Chart(pd.DataFrame({'Equal_Opp_Diff': [0]})).mark_rule().encode(y='Equal_Opp_Diff')
line2 = alt.Chart(pd.DataFrame({'Accuracy_Tradeoff': [0]})).mark_rule().encode(x='Accuracy_Tradeoff')

Equal_Opp = Equal_Opp + line + line2
Equal_Opp
```
